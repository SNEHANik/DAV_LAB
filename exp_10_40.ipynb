```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>D3.js Event Handling - Bar Chart</title>
    <script src="https://d3js.org/d3.v7.min.js"></script>
</head>
<body>

<!-- Create an SVG element -->
<svg width="400" height="200"></svg>

<script>
    // Sample data for the bar chart
    const data = [10, 20, 30, 40, 50];

    // Select the SVG element
    const svg = d3.select("svg");

    // Create scales for the bar chart
    const xScale = d3.scaleBand()
                    .domain(d3.range(data.length))
                    .range([0, 400])
                    .padding(0.1);

    const yScale = d3.scaleLinear()
                    .domain([0, d3.max(data)])
                    .range([200, 0]);

    // Create bars for the bar chart
    svg.selectAll("rect")
        .data(data)
        .enter()
        .append("rect")
        .attr("x", (d, i) => xScale(i))
        .attr("y", d => yScale(d))
        .attr("width", xScale.bandwidth())
        .attr("height", d => 200 - yScale(d))
        .style("fill", "steelblue")

        // Event handling: On mouseover, change the color of the bar
        .on("mouseover", function() {
            d3.select(this)
                .style("fill", "orange");
        })

        // Event handling: On mouseout, reset the color of the bar
        .on("mouseout", function() {
            d3.select(this)
                .style("fill", "steelblue");
        });
</script>

</body>
</html>
```
